### while loops is used when we don't know how many time the loops will iterate

```python
i=1 
while(i<=5):
    print(i)
    i+=1
```

```python
i=2
while(i<=50):
    if i % 2 == 0:
        print(i)
    i+=1
```

```python
i=2
while(i<=50):
    print(i)
    i+=2
```

```python
i=1
while(i<=50):
    if i % 2 != 0:
        print(i)
    i+=1
```

```python
num=int(input("Enter any number"))
i=1 
result=0
while(i<=10):
    result = num * i
    print(f"{num} x {i} = ", result)
    i+=1
```

```python
num=int(input("Enter a number")) 

i=1
count=0 
while(i<=num-1):
    if num % i == 0:
        count += 1
    i += 1
if count == 1:
    print("Prime")
else:
    print("not a prime")
```
