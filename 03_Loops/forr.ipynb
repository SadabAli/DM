### Loops is used to do iteration multiple times

```python
l=[1,2,3,2,1,2,3,2] 
for i in l:
    print(i)
```

```python
t='fjggiggi'
for i in t:
    print(i)
```

```python
t='I am Batman' 
for i in t.split():
    print(i)
```

```python
for i in range(2,20,2):
    print(i)
```
