```python
# Implements students attendence system using Set
student_db=[]
def Attendance():
    while True:
        roll=input("Enter your roll number (prss q for exit)")
        if roll.lower() == 'q':
            break
        student_db.append(int(roll))
    return set(student_db)
# print(Attendance())
```
