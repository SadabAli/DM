```python
l=[1,2,3,4,5,6] 

odd = list(filter(lambda x: x % 2 != 0,l))
print(odd)

even = list(filter(lambda x: x % 2 == 0,l))
print(even)
```

```python
# 2 list 1->prime, 2->non prime
l=[0,1,2,3,5,8,13] 

prime = list(filter(lambda x: x > 1 and all(x % i != 0 for i in range(2,int(x**.5) + 1)),l))
print(prime)
```
